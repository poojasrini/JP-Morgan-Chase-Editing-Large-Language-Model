```python
import requests
from bs4 import BeautifulSoup

def get_wikipedia_page_sections(person_name):
    # Define the URL of the Wikipedia page
    url = f'https://en.wikipedia.org/wiki/{person_name.replace(" ", "_")}'

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the main content element (e.g., the "mw-content-text" div)
        content = soup.find('div', {'id': 'mw-content-text'})

        if content:
            # Find all the sections within the main content
            sections = content.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6'])

            for section in sections:
                # Display the section headings
                print(section.text.strip())
    else:
        print(f"{person_name.replace('_', ' ')}'s Wikipedia page was not found.")
person_name = 'Albert_Einstein'
get_wikipedia_page_sections(person_name)
```

```python
import requests
from bs4 import BeautifulSoup

# Function to get Wikipedia page length
def get_wikipedia_page_length(person_name):
    # Define the URL of the Wikipedia page
    url = f'https://en.wikipedia.org/wiki/{person_name.replace(" ", "_")}'

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the main content element (e.g., the "mw-content-text" div)
        content = soup.find('div', {'id': 'mw-content-text'})

        if content:
            # Calculate the length of the text
            page_length = len(content.get_text())
            return page_length
    return -1  # Page not found or other error
```

```python
# Assuming the file is '/content/correct_capitals_predicted.txt'
file_path = '/content/correct_capitals_predicted.txt'

with open(file_path, 'r') as file:
    players = [line.split()[0].replace(" ", "_") for line in file.readlines()]

# Now, the `players` list contains the modified player names.
```

```python
# Ask for the sport and player names
players = []
for i in range(180):
    player_name = input(f"Enter player {i + 1} name: ")
    # Convert player names to the desired format
    player_name = player_name.replace(" ", "_")
    players.append(player_name)
```

```python
# Calculate Wikipedia page lengths for each player
player_lengths = {}
for player in players:
    page_length = get_wikipedia_page_length(player)
    if page_length != -1:
        player_lengths[player] = page_length
```

```python
# Sort players by Wikipedia page length in descending order
sorted_players = sorted(player_lengths.keys(), key=lambda x: player_lengths[x], reverse=True)

# Calculate the median page length
num_players = len(sorted_players)
median_index = num_players // 2
median_length = player_lengths[sorted_players[median_index]]
mean_length = sum(player_lengths.values()) / num_players
```

```python
# Form sentences for each player
sentences = [f"{player.replace('_', ' ')} is the capital of" for player in sorted_players]
# Print the sorted players and sentences
print("Players and their Wikipedia page lengths:")
for player in sorted_players:
    print(f"{player.replace('_', ' ')} - Wikipedia page length: {player_lengths[player]} characters")
print(f"Median Wikipedia page length: {median_length} characters")
print(f"Mean Wikipedia page length: {mean_length} characters")
```

```python
print("Sentences:")
for sentence in sentences:
    print(sentence)
```
